```python
!pip install umap-learn
from sklearn.decomposition import PCA
import umap

# Select only the numerical columns
numerical_cols = data.select_dtypes(include=['number']).columns
numerical_data = data[numerical_cols]

# Standardize the numerical data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numerical_data)

# ------------PCA----------
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Apply UMAP
reducer = umap.UMAP(n_components=2)
umap_result = reducer.fit_transform(scaled_data)

# Plot the PCA result
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=data['Gender'].apply(lambda x: 1 if x == 'Male' else 0), cmap='viridis', s=50)
plt.title('PCA Result')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')

# Plot the UMAP result
plt.subplot(1, 2, 2)
plt.scatter(umap_result[:, 0], umap_result[:, 1], c=data['Gender'].apply(lambda x: 1 if x == 'Male' else 0), cmap='viridis', s=50)
plt.title('UMAP Result')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')

plt.show()
```

Based on the visualizations, **UMAP worked better than PCA** for dimensionality reduction in this dataset. UMAP provided a clearer separation of clusters, capturing the underlying structure of the data more effectively, whereas PCA showed more overlap and less distinct groupings. This indicates that UMAP's non-linear approach is more suited for visualizing and distinguishing complex patterns.

### Clustering

#### Elbow Method

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Define the range of clusters to try
cluster_range = range(2, 11)  # Typically, start with a range of 2 to 10 or more
sse = []  # Sum of squared distances
silhouette_scores = []

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_scaled)
    sse.append(kmeans.inertia_)  # Inertia: Sum of squared distances to closest cluster center
    silhouette_avg = silhouette_score(features_scaled, kmeans.labels_)
    silhouette_scores.append(silhouette_avg)

# Plotting the Elbow Method
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(cluster_range, sse, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal k')

# Plotting the Silhouette Scores
plt.subplot(1, 2, 2)
plt.plot(cluster_range, silhouette_scores, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Optimal k')

plt.tight_layout()
plt.show()
```

#### K means experimentation

```python
# The plot shows a noticeable “elbow” at k=4

# Apply K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(features_scaled)

# Add the cluster labels to the original dataframe
data['KMeans_Labels'] = kmeans_labels

# Visualize the K-Means clustering result
plt.figure(figsize=(10, 6))
sns.scatterplot(x=features_scaled[:, 1], y=features_scaled[:, 2], hue=kmeans_labels, palette='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
# The silhouette score is highest at k=6
# Apply K-Means
kmeans = KMeans(n_clusters=6, random_state=42)
kmeans_labels = kmeans.fit_predict(features_scaled)

# Add the cluster labels to the original dataframe
data['KMeans_Labels'] = kmeans_labels

# Visualize the K-Means clustering result
plt.figure(figsize=(10, 6))
sns.scatterplot(x=features_scaled[:, 1], y=features_scaled[:, 2], hue=kmeans_labels, palette='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

# EXTRA: How do clustering and modeling compare? What are the advantages of each? Why would you want to use one over the other?

Clustering and modeling are both fundamental techniques, with **clustering** being an unsupervised learning approach that **groups data points based on similarity without predefined label**s, and **modeling**typically referring to supervised learning where models are **trained on labeled data to make predictions**.



**Clustering is advantageous for discovering hidden patterns or natural groupings in data**, which is particularly useful for exploratory data analysis, customer segmentation, and anomaly detection.


**Modeling**, excels in tasks where the goal is to **predict an outcome based on known input-output relationships**, such as classification or regression.


The choice between clustering and modeling depends on the availability of labeled data and the specific objective: use **clustering for exploratory insights and unsupervised pattern detection**, and **modeling for predictive tasks where labeled data is available**.
